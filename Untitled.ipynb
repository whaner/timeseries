```python
#Importing the dataset
df = pd.read_csv("passenger_entrance_2022.csv")
```

```python
#Transforming the variable type
df['month'] = pd.to_datetime(df['month'], format='%Y-%m')
print(df.head())
```

```python
df.info()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
#Setting the month feature as index
df.index = df['month']
del df['month']
print(df.head())
```

```python
# Release Date from the index
df = df.reset_index()
# Change data from the wide format to the long format
df = pd.melt(df, id_vars='month', value_vars=['MDU'])
df.columns = ['ds', 'station', 'y']
df.head()
```
